```python
import sympy as sp
```

```python
unidades_inversas =[]
unidades = []
for a in range(26):
    try:
        unidades_inversas.append(sp.invert(a,26))
        unidades.append(a)
    except:
        continue
        
print(unidades)
print(unidades_inversas) #lista de unidades
```

```python
import sympy.crypto.crypto as spc
```

```python
# vamos a buscar como se cifró
# usa este código para encontrar como se cifro (la llave)

text = "THE GOOD"
for a in unidades:
    print(spc.encipher_affine(text,(a,1)))
```

```python
## cuado encontremos como se cifró encotramos como se cifro

text = "XBCKQ QLXBC DMLMZ LXBCO KRE"
a = 21 #esto se meodifica
b = 1 # esto se modifica
spc.decipher_affine(text,(a,b))
```
